```python
import json
movies_data = movies
def parse_col_json(column, key):
    """
    Args:
        column: string
            name of the column to be processed.
        key: string
            name of the dictionary key which needs to be extracted
    """
    for index,i in zip(movies_data.index,movies_data[column].apply(json.loads)):
        list1=[]
        
        for j in range(len(i)):
            list1.append((i[j][key]))# the key 'name' contains the name of the genre
        movies_data.loc[index,column]=str(list1)
    
parse_col_json('genres', 'name')
parse_col_json('production_countries', 'name')
parse_col_json('spoken_languages', 'name')
#parse_col_json('cast', 'name')
```

# all strip(),split(),replace() works with a inverted coma inside the brackets

```python
movies['genres']=movies['genres'].str.strip('[]').str.replace(' ','').str.replace("'",'')
movies['genres']=movies['genres'].str.split(',')
movies.head()
```

```python
list = []
for i in movies["genres"]:
    list.extend(i)     
list    
```
