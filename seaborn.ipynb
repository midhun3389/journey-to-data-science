```python
import seaborn as sns
import pandas as pd
iris_data =pd.read_csv("iris.csv")
g = iris_data
sns.FacetGrid(g,col = "Species",hue = "Species",size = 5)\
.map(plt.hist,"PetalLengthCm")\
.add_legend()
```

```python
sns.boxplot(x = "Species", y = "PetalWidthCm", data = g)
```

```python

sns.FacetGrid(iris_data,col = "Species", hue = "Species",size =5)\
.map(plt.scatter,"PetalLengthCm", "PetalWidthCm")\
.add_legend()
```

```python

sns.boxplot(x = "Species", y = "PetalWidthCm", data = iris_data)
sns.stripplot(x = "Species", y = "PetalWidthCm", data = iris_data, jitter= False , edgecolor="yellow")
```
