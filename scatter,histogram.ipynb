```python
from matplotlib import pyplot as plt
import pandas as pd

x = [1,2,3,4,5,6,7,8,9]
y = [1,1,2,2,3,4,5,5,6]
z = [1,2,3,4,5,6,8,9,5]
```

```python
plt.plot(x,y)
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend(["x","y"])
plt.show()
```

```python
plt.plot(x,y,marker = 'o',linestyle = "--",color = "r", label = "y")
plt.plot(x,z,marker = 'o',linestyle = "-",color = "g", label = "z")

plt.xlabel("x axis")
plt.ylabel("y axis")
plt.legend(["y","z"])
plt.show()
```
