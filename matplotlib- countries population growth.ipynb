```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
population = pd.read_csv("countries.csv")
population
```

```python
population.country == "India"
```

```python
India = population[population.country == "India"]
India
```

```python
China = population[population.country == "China"]
China
```

```python
us = population[population.country == "United States"]
us
```

```python
plt.plot(India.year,India.population)
#plt.plot(India.year,India.population/10**6)  this divides

plt.plot(China.year,China.population)

plt.plot(us.year,us.population)
plt.title("population growth")
plt.legend(["India","China","us"])
plt.xlabel("year")
plt.ylabel("population")

plt.show()
```
