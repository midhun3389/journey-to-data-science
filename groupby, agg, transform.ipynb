```python
a =[["a"],["g"]]
index = pd.MultiIndex.from_arrays(a,names = ["first","second"])
data = pd.DataFrame({"a":["1"]},index = index)
data
```

```python
import numpy as np
a =[["a","b","b","c","d","d"],["g","f","t","r","e","o"]]
index = pd.MultiIndex.from_arrays(a,names = ["first","second"])
data = pd.DataFrame({"A":range(1,7),"B": range(8,14),"C":np.arange(2,8)},index = index)
data
```

# groupby

```python
data.groupby("first").sum().add_prefix("total.")
```

```python
data.groupby([pd.Grouper(level = 0)]).sum().add_prefix("total.")
```

```python
datas = data.groupby([pd.Grouper(level = 1)]).sum().add_prefix("total.")
datas
```

```python
datas["total.A"].sum()
```

```python
datas = data.groupby([pd.Grouper(level = "first")]).sum().add_prefix("total.")
datas
```

```python
grouped = data.groupby("A")
grouped
```

```python
import pandas as pd
grouped.aggregate([np.mean,np.median])
```
